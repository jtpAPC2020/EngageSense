## Flip the images in disgust

```python
# Specify input and output directories
input_directory = 'FER2013 full/train/disgust'
output_directory = 'FER_Augmented-Image/disgust'

# Ensure output directory exists
os.makedirs(output_directory, exist_ok=True)

# List all image files in the input directory
image_files = [file for file in os.listdir(input_directory) if file.lower().endswith(('.jpg', '.png', '.jpeg'))]

# Loop through each image file
for image_file in image_files:
    # Load the image using cv2
    image_path = os.path.join(input_directory, image_file)
    img = cv2.imread(image_path)
    
    # Apply horizontal flip using cv2
    horiz_flipped_img = np.fliplr(img)
    
    # Save the flipped image to the output directory using cv2
    output_path = os.path.join(output_directory, 'flipped_' + image_file)
    cv2.imwrite(output_path, horiz_flipped_img)

print("Horizontal flipping and saving complete.")
```

## Rotate disgust

```python
# Specify input and output directories
input_directory = 'FER2013 full/train/disgust'
output_directory = 'FER_Augmented-Image/disgust/rotated'

# Ensure output directory exists
os.makedirs(output_directory, exist_ok=True)

# List all image files in the input directory
image_files = [file for file in os.listdir(input_directory) if file.lower().endswith(('.jpg', '.png', '.jpeg'))]

# Loop through each image file
for image_file in image_files:
    # Load the image using cv2
    image_path = os.path.join(input_directory, image_file)
    img = cv2.imread(image_path)
    
    # Apply rotation using skimage.transform.rotate
    rotated_angle = np.random.uniform(-10, 10)  # Random rotation angle between -10 and 10 degrees
    rotated_img = transform.rotate(img, rotated_angle)
    
    # Normalize pixel values to [0, 255]
    rotated_img = (rotated_img * 255).astype(np.uint8)
    
    # Find the black pixels (zero values) and set them to white
    rotated_img[rotated_img == 0] = 255
    
    # Save the rotated image to the output directory using cv2
    output_path = os.path.join(output_directory, 'rotated_' + image_file)
    cv2.imwrite(output_path, rotated_img)
    
print("Image rotation and saving complete.")
```

## Flip the images in angry, fear, neutral, sad and surprise

```python
# List of folder names
folder_names = ['angry', 'fear', 'neutral', 'sad', 'surprise']

# Loop through each folder
for folder_name in folder_names:
    # Specify input and output directories
    input_directory = os.path.join('FER2013 full/train', folder_name)
    output_directory = os.path.join('FER_Augmented-Image', folder_name)
    
    # Ensure output directory exists
    os.makedirs(output_directory, exist_ok=True)
    
    # List all image files in the input directory
    image_files = [file for file in os.listdir(input_directory) if file.lower().endswith(('.jpg', '.png', '.jpeg'))]
    
    # Loop through each image file
    for image_file in image_files:
        # Load the image using cv2
        image_path = os.path.join(input_directory, image_file)
        img = cv2.imread(image_path)
        
        # Apply horizontal flip using cv2
        horiz_flipped_img = np.fliplr(img)
        
        # Save the flipped image to the output directory using cv2
        output_path = os.path.join(output_directory, 'flipped_' + image_file)
        cv2.imwrite(output_path, horiz_flipped_img)
    
    print(f"Horizontal flipping and saving complete for {folder_name} folder.")
```

## Adjust brightness and contrast to disgust images

```python
# Specify input and output directories
input_directory = 'FER2013 full/train/disgust'
output_directory = 'FER_Augmented-Image/disgust/brightness'

# Ensure output directory exists
os.makedirs(output_directory, exist_ok=True)

# List all image files in the input directory
image_files = [file for file in os.listdir(input_directory) if file.lower().endswith(('.jpg', '.png', '.jpeg'))]

# Loop through each image file
for image_file in image_files:
    # Load the image using cv2
    image_path = os.path.join(input_directory, image_file)
    img = cv2.imread(image_path)
    
    # Increase brightness and contrast using cv2
    alpha = .5  # Controls contrast
    beta = 10    # Controls brightness
    adjusted_image = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
    
    # Save the adjusted image to the output directory using cv2
    output_path = os.path.join(output_directory, 'adjusted_' + image_file)
    cv2.imwrite(output_path, adjusted_image)

print("Brightness and contrast adjustment and saving complete.")
```

```python
# Specify input and output directories
input_directory = 'FER2013 full/train/disgust'
output_directory = 'FER_Augmented-Image/disgust/brightness'

# Ensure output directory exists
os.makedirs(output_directory, exist_ok=True)

# List all image files in the input directory
image_files = [file for file in os.listdir(input_directory) if file.lower().endswith(('.jpg', '.png', '.jpeg'))]

# Initialize the count for adjusted images
adjusted_count = 0

# Loop through each image file
for image_file in image_files:
    # Load the image using cv2
    image_path = os.path.join(input_directory, image_file)
    img = cv2.imread(image_path)
    
    # Compute average brightness of the image
    avg_brightness = np.mean(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))
    
    # Define a threshold for low brightness
    brightness_threshold = 100  # Adjust this threshold as needed
    
    # If average brightness is below the threshold, adjust brightness and contrast
    if avg_brightness < brightness_threshold:
        alpha = 1.5  # Controls contrast
        beta = 40    # Controls brightness
        adjusted_image = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
        adjusted_count += 1
        
        # Save the adjusted image to the output directory using cv2
        output_path = os.path.join(output_directory, 'adjusted_' + image_file)
        cv2.imwrite(output_path, adjusted_image)

# Print the count of adjusted images
print(f"Adjusted {adjusted_count} images.")
print("Brightness and contrast adjustment and saving complete.")
```
